```python
from pathlib import Path

efficient = Path("../data/raw/efficient")
list(efficient.glob("*"))
```

# Find the best representation for data

```python
lines = (efficient / "monster.txt").read_text().split("\n")
res = []
for line in lines:
    res.append(dict(zip(["name", "map", "town"], line.split("\t"))))
monsters = res
monsters
```

```python
lines = (efficient / "tour.txt").read_text().split("\n")
lines[:10]
tour = lines
tour
```

```python
from difflib import get_close_matches

tour.index(get_close_matches(monsters[15]["map"], tour)[0])
```

```python
res = []
for monster in monsters:
    matches = get_close_matches(monster["map"], tour)
    if not matches:
        if monster["name"] == "Sand Rat":
            res.append(tour.index("Sahel 3"))
        else:
            res.append(999)
    else:
        res.append(tour.index(matches[0]))
res
tour_order = res
```

```python
for i, order in enumerate(tour_order):
    monsters[i]["tour_order"] = order
```

```python
import json

output = Path("../data/processed/efficient")
output.mkdir(parents=True, exist_ok=True)
json.dump(monsters, (output / "monsters.json").open("w"), indent=2)
```
